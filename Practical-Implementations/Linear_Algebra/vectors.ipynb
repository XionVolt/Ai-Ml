```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the new vectors
v = np.array([1, 1])
w = np.array([1, 1])
scalar1 = 11
scalar2 = -10
v*=scalar1
w*=scalar2


# Plot the vectors
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r')
plt.quiver(0, 0, w[0], w[1], angles='xy', scale_units='xy', scale=1, color='b')

# Set the limits of the plot
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.gca().set_aspect('equal', adjustable='box')

# Add grid and labels
plt.grid()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Math - Linear Combinations')

# Show the plot
plt.show()
```

```python
# Calculate the resultant vector
resultant_vector = v + w

# Plot the resultant vector
plt.quiver(0, 0, resultant_vector[0], resultant_vector[1], angles='xy', scale_units='xy', scale=1, color='g')

# Set the limits of the plot
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.gca().set_aspect('equal', adjustable='box')

# Add grid and labels
plt.grid()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('resultant_vector = v + w')

# Show the plot
plt.show()
```

```python
# demonstrate all vector multiples(here of vector u) lies on the same line 
u = np.array([2, 4])
first_mul= np.array([1, 2])
second_mul= np.array([3, 6])
third_mul= np.array([-10, -20])
fourth_mul= np.array([-2, -4])

# Plot the resultant vector
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='g')
plt.quiver(0, 0, first_mul[0], first_mul[1], angles='xy', scale_units='xy', scale=1, color='r')
plt.quiver(0, 0, second_mul[0], second_mul[1], angles='xy', scale_units='xy', scale=1, color='b')
plt.quiver(0, 0, third_mul[0], third_mul[1], angles='xy', scale_units='xy', scale=1, color='y')
plt.quiver(0, 0, fourth_mul[0], fourth_mul[1], angles='xy', scale_units='xy', scale=1, color='m')

# Set the limits of the plot
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.gca().set_aspect('equal', adjustable='box')

# Add grid and labels
plt.grid()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Vector u and its multiples vectors')

# Show the plot
plt.show()
```

```python
# How to find distance between two vectors ? By finding the Eucledian distance of vector a - b 
# Eucledian distance between two vectors
a = np.array([2, 7])
b = np.array([6, 2])
difference_Vector = a-b
# Plottin the vectors
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='r')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='b')
plt.quiver(b[0], b[1], difference_Vector[0], difference_Vector[1], angles='xy', scale_units='xy', scale=1, color='g')

# Add dashed lines


# right angle triangle , means we can also use pythagoras theorem to find the distance between two vectors (a and b)
plt.plot([2, 6], [2, 2], 'k--')  # [2,6] means on x-axis from 2 to 6 , [2,2] means on y-axis from 2 to 2
plt.plot([2, 2], [2,7], 'k--')  # [2,2] means on x-axis from 2 to 2 , [2,7] means on y-axis from 2 to 7


# Set the ticks on the axes , ticks are the numbers and their corresponding lines on the axis , ticks helps to set distance between the columns/rows on the axis
plt.xticks(np.arange(-1, 8, 1)) # means from -1 to 7(excluding 8 as it takes 1 less) we get grid numbers and their corresponding lines (with gap between of 1) on x-axis (between columns)
plt.yticks(np.arange(-1, 8, 1)) # means from -1 to 7(excluding 8 as it takes 1 less) we get grid numbers and their corresponding lines (with gap between of 1) on y-axis (between rows)

# Set the limits of the plot , labels just make the axis rows/columns span given range , they not add grid lines and numbers like ticks , so labels are kind of defining size of chart
plt.xlim(-2, 8) 
plt.ylim(-2, 8)

plt.gca().set_aspect('equal', adjustable='box')

# Add grid and labels
plt.grid()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Vector a, b and difference vector')
# Show the plot
plt.show()
```

```python
# Eucledian of a difference_Vector (that is distance between `a` and `b` vectors tip)
# np.linalg.norm() is used to find the eucledian distance of a vector
eucledian_distance = np.linalg.norm(difference_Vector)
eucledian_distance
```

##### Now observe above cell  , the distance between a and b vectors and the value of eucledian_distance is same because tail of the difference_Vector is at the tip of Vector b and tip of the difference_Vector is at tip of Vector a. And We know eucledian distance of the vector is the distance between the tip and tail of that vector .
