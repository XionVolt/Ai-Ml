```python
# plotting a simpler function 
price = [711,92,13,54,5,2500] # independent variables
year = [2010,2011,2012,2013,2014,2025] # dependent variables
    #  x(genrally we plot categorical column on x axis) , y values
plt.plot(year,price)
plt.show()
```

```python
import pandas as pd

data = {
    'Year': [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016,2017],
    'RG Sharma': [404, 362, 404, 372, 433, 538, 390, 482, 489,333],
    'V Kohli': [165, 246, 307, 557, 364, 639, 359, 505, 973,308]
}

df1 = pd.DataFrame(data)
df1
```

```python
# v.kohli performance in a years
# we can also change graph styles

plt.style.use('classic') # other values like ggplot, dark_background, etc.

plt.plot(df1['Year'],df1['V Kohli'])
plt.show()
# see below in x axis is you see 2e3 it means 2*10^3 which is 2000 so its like 2000 + 8 = 2008 , 2000 + 9 = 2009 and so on ...
```

```python
# plotting multiple plots -> https://www.youtube.com/live/XaKn_cKFlSY?si=cd_AmOyh9owqhFFK&t=1407

plt.style.use('default')
plt.plot(df1['Year'],df1['V Kohli'])
plt.plot(df1['Year'],df1['RG Sharma'])
```

```python
# labels title and axis, and columns label (using colors)

plt.plot(df1['Year'],df1['V Kohli'])
plt.plot(df1['Year'],df1['RG Sharma'])
plt.title('V Kohli vs RG Sharma')

plt.legend(['V Kohli', 'RG Sharma'])
plt.xlabel('Year')
plt.ylabel('Runs')
```

```python
# color (hex), line (width and style) and marker (size) -> https://www.youtube.com/live/XaKn_cKFlSY?si=gAJ6TM_EpbdZJd5-&t=1737

plt.plot(df1['Year'],df1['V Kohli'],color='red',linestyle='--',linewidth=3)
plt.plot(df1['Year'],df1['RG Sharma'],color='blue', linestyle='dashdot')

plt.title('V Kohli vs RG Sharma')

plt.legend(['V Kohli', 'RG Sharma'])

plt.xlabel('Year')
plt.ylabel('Runs')

plt.show()
```

```python
# markers (what are markers, and their different types -> https://www.youtube.com/live/XaKn_cKFlSY?si=Q-SQp7Ru-A6ifCPM&t=1957)

plt.plot(df1['Year'],df1['V Kohli'],color='red',linestyle='--',marker='o',markersize=10) 
plt.plot(df1['Year'],df1['RG Sharma'],color='blue', linestyle='dotted',marker='+')

plt.title('V Kohli vs RG Sharma')

plt.legend(['V Kohli', 'RG Sharma']) # so according to linestyle and marker and markersize, legend also modified itself

plt.xlabel('Year')
plt.ylabel('Runs')

plt.show()
```

```python
# give name to labels inside legend, and change legend location

# markers (what are markers, and their different types -> https://www.youtube.com/live/XaKn_cKFlSY?si=Q-SQp7Ru-A6ifCPM&t=1957)

plt.plot(df1['Year'],df1['V Kohli'],color='red',linestyle='--',marker='o',markersize=10,label='V Kohli') # so you can also give names with label param
plt.plot(df1['Year'],df1['RG Sharma'],color='blue', linestyle='dotted',marker='>')

plt.title('V Kohli vs RG Sharma')

plt.legend(['V Kohli', 'RG Sharma'],loc='center') # so according to linestyle and marker and markersize, legend also modified itself, default is `best`

plt.xlabel('Year')
plt.ylabel('Runs')

plt.show()
```

```python
# limiting axes, and turning grid on -> https://www.youtube.com/live/XaKn_cKFlSY?si=zcPcIRoOuob-Sy-k&t=2327
price = [711,92,13,54,5,442500] # independent variables
year = [2010,2011,2012,2013,2014,2025] # dependent variables

# outlier comes so other values are not visible

plt.plot(year,price,color='red',linestyle='solid',marker='D',markersize=10)

plt.xlabel('Year')
plt.ylabel('Price')


plt.ylim(0, 2000) # y-axis limit from 0 to 2000, so we can limit y axis to see small variations against big outliers
plt.xlim(2009, 2015)

plt.title('Stock Price of a Company')

plt.legend(['Stock Price'])

plt.grid()

plt.show() 
```

## [Bar chart](https://www.youtube.com/live/XaKn_cKFlSY?si=cLruYZfJV6otNfdF&t=3827)

- Can be used for both both bivariate and univarite analysis

```python
# simple bar chart
tips
```

```python
avg_tips = tips.groupby('sex')['tip'].mean()

plt.bar(avg_tips.index, avg_tips.values)
plt.ylabel('Average Tip')
plt.title('Average Tip by Sex')
plt.show()
```

```python
# let's say we wanna find average `total_bill` of male smoker and no smoker, female smoker and female non-smoker

tips.groupby(['sex','smoker'])['total_bill'].mean().unstack()
```

```python
msfs = tips.groupby(['sex','smoker'])['total_bill'].mean()
```

```python
fig, ax = plt.subplots()

msfsbar = pd.Series(index=msfs.index.map(lambda x: x[0] + ' ' + 'Smoker  ' if x[1] == 'Yes' else x[0] + ' ' + 'No-Smoker'), data = msfs.values)
bars = plt.bar(msfsbar.index, msfsbar.values,color=['#01084f','#391954','#631e50','#a73c5a'],width=0.7)
ax.bar_label(bars, msfsbar.values.round(2))
# so we can also give different colors to each bar
```

```python
# horizontal bar chart -> https://www.youtube.com/live/XaKn_cKFlSY?si=xqYDhmvfMp8MJYF3&t=4047

fig, ax = plt.subplots()

msfsbar = pd.Series(index=msfs.index.map(lambda x: x[0] + ' ' + 'Smoker  ' if x[1] == 'Yes' else x[0] + ' ' + 'No-Smoker'),
 data = msfs.values)
bars = plt.barh(msfsbar.index, msfsbar.values,color=['#01084f','#391954','#631e50','#a73c5a'],height=0.7)
ax.bar_label(bars, msfsbar.values.round(2))
# so we can also give different colors to each bar

# Tip: use when x axis becomes too long
```

# [Histogram](https://www.youtube.com/live/XaKn_cKFlSY?si=4rh3U6l0BjM1U-L0&t=5707)

- Used in univariate analysis
- Use case: Frequency count

```python
# simple data 0> https://www.youtube.com/live/XaKn_cKFlSY?si=zgflOpA9YBr3ZacF&t=6230
dataFlowersSize = [77,41,21,32,55]

# in bins in 0-50(50 is exclusive) and 50-100(100 is exclusive)
plt.hist(dataFlowersSize, bins=np.arange(0,101,50))
```

```python
# example of hist on some more data -> https://www.youtube.com/live/XaKn_cKFlSY?si=AiVjE1MrNIM6DXh4&t=5937
```

```python
# to give hists a labels of their corresponding y-axis value
fig,ax = plt.subplots()

# `counts` are y-axis(frequency of each bin) values, bins are that we give in `bins` param, the `patches` are bars
counts, bins, patches = plt.hist(dataFlowersSize, bins=np.arange(0,101,50))


plt.xlabel('Flowers Size')
plt.ylabel('Frequency')

ax.bar_label(patches, counts)
```

```python
# logarithmic scale(especially useful for big data, when large bins hide small ones)  -> https://www.youtube.com/live/XaKn_cKFlSY?si=nA7k6mwd6dhiNXEP&t=6127

# so in log scale we, so what it essentially do , it treating 0-2 and 2-4 as equal 
plt.hist(dataFlowersSize, bins=np.arange(0,101,50), log=True)
```

# [pie chart](https://www.youtube.com/live/XaKn_cKFlSY?si=xz_2EaevG16ifde3&t=6357)
- Univariate/Bivariate analysis
- Categorical vs Numerical
-  Use case -> To find contribution on a standard scale(standard scale means 100 percent)

```python
# simple
pieData1 = [16,17,77,88,20,27]
stocks = ['Amazon', 'Google', 'Facebook', 'Nivida', 'Microsoft', 'Apple']
plt.pie(pieData1,labels=stocks,startangle=90,autopct='%1.1f%%') # means show percentage of each value, to 1 decimal place
plt.title('Stocks Distributions inside CSET50 ETF')
```

```python
plt.pie(pieData1,labels=stocks,startangle=90,autopct='%1.1f%%', explode=[0.1,0,0,0,0,0]) # explode is to make a gap in pie(often use to highlight some specific value)
```

```python
pieData1
```

```python
# you can also assign custom function to `autopct` param
plt.figure(figsize=(15,12))
def show_pct(percent, allvalues):
    absolute = round((percent/100)*np.sum(allvalues)) # give us number from percentage, back
    return "{:.1f}%\n({} shares)".format(percent, absolute)

plt.pie(pieData1,labels=stocks,startangle=90,autopct=lambda percent: show_pct(percent, pieData1)) 
plt.show()
```
