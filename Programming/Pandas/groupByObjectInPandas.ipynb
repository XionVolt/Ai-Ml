```python
# group by always apply on categorical columns
import pandas as pd
import numpy as np

movies = pd.read_csv('imdb-top-1000.csv')
movies
```

```python
# understand this example -> https://www.youtube.com/live/LPBjF4_gZnI?si=-KFPZHtL6IwZj94e&t=557

genres = movies.groupby('Genre') # so basically we know genres column is a categorical column, and therefore panadas classifies each genre in a separate related groups
genres # so it basically a dataframe groupby object, we can also make series group by object -> https://www.youtube.com/live/LPBjF4_gZnI?si=t-n6AzxqEfbY_zdW&t=1427 
```

```python
# Appply a aggregate (builtin functions like add, min, max, mean, etc.) function to each group
genres.sum()
```

```python
genres.min() # will give min value for each column for each genre
```

```python
# but the best approach is first index the column you want to perform aggregation on, and then apply the aggregate function(it will save computation by only performing aggregation on that column)
genres.max()
```

```python
genres['Gross'].std()
```

```python
# find the top5 genres by total earning, by dataframe groupby object
genres['Gross'].sum().sort_values(ascending=False).head(5)
```

```python
# find the genre with highest average imdb rating

genres['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
```

```python
# find director with most imdb movies 
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(1)
```

```python
# find the no. of movies by each actor
movies.groupby('Star1').size().sort_values(ascending=False).head(5)

# or like this without groupBy() -> movies['Star1'].value_counts()
```
