````python
# group by always apply on categorical columns
import pandas as pd
import numpy as np

movies = pd.read_csv('imdb-top-1000.csv')
movies
````

````python
# understand this example -> https://www.youtube.com/live/LPBjF4_gZnI?si=-KFPZHtL6IwZj94e&t=557

# we can also loop on these groupby objects, that we will see later

genres = movies.groupby('Genre') # so basically we know genres column is a categorical column, and therefore pandas classifies each genre in a separate related groups
print(genres) # so it basically a dataframe groupby object, we can also make series group by object -> https://www.youtube.com/live/LPBjF4_gZnI?si=t-n6AzxqEfbY_zdW&t=1427 
len(genres) # total 14 groups have been formed, means we can say pandas create 14 separate groups on the basis of 'Genre' columns
""" or use movies['Genre].nunique(), which will give us number of unique values present in Genre column"""
# movies['Genre'].nunique()
````

````python
# Appply a aggregate (builtin functions like add, min, max, mean, etc.) function to each group
genres.sum() # so basically it sum all numerical values in each column for each group(genre here), and for non-numerical columns, it cocatenate the values
````

````python
# as_index=False, will give us a dataframe with column names as Star1 and size as new columns instead of treating those two as index
movies.groupby(['Star1','Genre' ]).count() # so you can also group by multiple columns
# ('Aamir Khan', 'Action'), here count is 2 means Aamir Khan has 2 movies in Action genre
````

````python
genres.min() # will give min value for each column for each genre
````

````python
""" but the best approach is first index the column you want to perform aggregation on, and then apply the aggregate
function(it will save computation by only performing aggregation on that column only instead of all columns) """
genres.max()
````

````python
genres['Gross'].std() # calculate standard deviation for each genre of gross column
````

````python
# find the top5 genres by total earning, by dataframe groupby object, so `genres['Gross'].sum()` will do sum of all rows of Gross columns for each group()
genres['Gross'].sum().sort_values(ascending=False).head(5)
````

````python
# find the genre with highest average imdb rating, so `movies.groupby('Genre')['IMDB_Rating'].mean()` will do mean of all rows of IMDB_Rating columns for each group()
movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
````

````python
# find director with most imdb movies, so `movies.groupby('Director')['No_of_Votes'].sum()` will do sum of all rows of No_of_Votes columns for each group()
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(1)
````

````python
# find the higest rated movies of each genre -> we will see how to solve this with groupby later

# for now we can solve this simply, in this way
movies.sort_values(by='IMDB_Rating',ascending=False).drop_duplicates(subset='Genre') 
````

````python
# `first` and `last`(vice-versa) function on groupBy object
movies.groupby('Genre').first() # it will give first row of each genre, i.e., first movie details for each genre
````

````python
# but we have nth also 
movies.groupby('Genre').nth(6) # it will give 7th row of each genre, i.e., 7th movie details for each genre 
# but if you observe here, there are certain groups who don't have 7th movie, this is because some genres have less than 7 movies
````

````python
# get_group -> https://www.youtube.com/live/LPBjF4_gZnI?si=_Vi1_ZaLBRXy5HXN&t=2587
movies.groupby('Genre').get_group('Animation') # it will give all rows of Animation genre
# so basically get_group() is used to get the specific group based on the groupby object
````

````python
movies[movies['Genre'] == 'Animation'] 
""" second way without groupBy, but above method is fast, because groupby object is pre-computed,
which improves performance for large datasets """
````

````python
# groups explained -> https://www.youtube.com/live/LPBjF4_gZnI?si=QGHTI-4vreI_Golb&t=2747

movies.groupby('Genre').groups # this will give us a dictionary where keys are genre names and values are indexes of rows that belong to that genre
# movies.iloc[movies.groupby('Genre').groups['Animation']] then in this way you can fetch all movies related to specifc genre
````

````python
# describe -> https://www.youtube.com/live/LPBjF4_gZnI?si=nJKIo_RDHCSFBhaS&t=2857
movies.groupby('Genre').describe() # so this basically gives statistical summary for each group(genre here) of each numerical column(here gross, imdb_rating, no_of_votes)
````

````python
# sample -> https://www.youtube.com/live/LPBjF4_gZnI?si=nKJpy3AooXe443Pk&t=3001
movies.groupby('Genre').sample(frac=0.1) # this will give you 10% of rows randomly from each group(which is genre in this case)

# movies.groupby('Genre').sample(2, replace=True) 
# and replace=True means if there are less than 2 rows in a group, it will replace them with same rows again and again instead of raising a error
````

````python
# nunique -> https://www.youtube.com/live/LPBjF4_gZnI?si=2OdVS75KJsJkkN4h&t=3087
movies.groupby('Genre').nunique() # nunique will give you number of unique values present(for each column) in each group
````

````python
# agg method -> https://www.youtube.com/live/LPBjF4_gZnI?si=_I6OHIEMhnLGvAtZ&t=3327
# passing list

movies.groupby('Genre').agg(['min', 'count','max']) # it will give you min, count and max of each numerical column for each group genre here


# passing dict, so you can apply different aggregation functions to different columns, you can do this thing in dataframe too
genres.agg(
    { 
    'Gross': ['mean','sum'], 
    'IMDB_Rating': 'mean', 
    'No_of_Votes': 'sum',
    'Metascore': 'mean'
    }
) 
````

````python
# looping on groups -> https://www.youtube.com/live/LPBjF4_gZnI?si=A0amEHJQRIUKE3Bz&t=3887
#  genre names comes in genre, and in group the dataframe comes , of all columns and all rows(that comes in that group)

for genre, group in movies.groupby('Genre'):
    print(genre,'\n')
    print(group)
    print('----------------------------------------------','\n')
````

````python
# now let's find highest rated movies of each genre using groupby

for genre, group in movies.groupby('Genre'):
    print('Highest rated movie of',genre,'\n')
    print(group.nlargest(1, 'IMDB_Rating')['Series_Title'])
    print('----------------------------------------------','\n')
````

````python
# Note: if you return each group from function that you give in apply, it will give you original dataframe(in which that groupby is applied) 
def group_apply_example(group):
    return group
        

    

movies.groupby('Genre').apply(group_apply_example) # so its kind of concatente dataframes of each group
````

````python

# split(apply) combine 
# apply (builtin-method)
# explain: https://www.youtube.com/live/LPBjF4_gZnI?si=a1RGRsHaI-4tGoO4&t=4397 

# find the number of movies starting with A for each group -> https://www.youtube.com/live/LPBjF4_gZnI?si=H7DZJ9FS8QV4B7tA&t=4577
def isMoviesStartWithA(series_titles):
    # in series_title the series comes(values as Series_Title) of each group
    return series_titles.str.startswith('A').sum() # return for each group the count of movies in that group whose titles start with 'A'


movies.groupby('Genre')['Series_Title'].apply(isMoviesStartWithA)
# so in `x` param of this lambda function, each time comes a group, means `Series_Title` column for that group
````

````python
movies[movies['Series_Title'].apply(lambda x: x.startswith('A'))]['Genre'].value_counts() # this will give you count of movies starting with 'A' for each genre # this will give you count of movies starting with 'A' for each genre
# so this can also be done using this way without groupby, but it will be slower and less efficient for large datasets
````

````python
# find ranking of each movie in the group according to IMDB_Rating
# explanation -> https://www.youtube.com/live/LPBjF4_gZnI?si=8i2jdoXmIDQdCXMS&t=4907 this is the one more way

movies2 = movies.copy()
movies2['Rank'] = ''

newg = movies.groupby('Genre')['IMDB_Rating']
def rank_movies(gr):
    # Note : gr.index means all rows(that goes in the group) the original index of those rows
    movies2.loc[gr.index, 'Rank'] = gr.rank(ascending=False).astype(float)

newg.apply(rank_movies)  # this will add a new column 'Rank' to movies dataframe with relative rank of each movie(row) to its genre siblings
movies2
    
````

Normalization formula
```math

\frac{x - x_{min}}{x_{max} - x_{min}}

````python
# find normalized IMDB_Rating group wise
# normalization explained and how to do above question -> https://www.youtube.com/live/LPBjF4_gZnI?si=s9aax5kzhsdk278e&t=5157

def normalize_imdb_rating(group):
    min = group['IMDB_Rating'].min()
    max = group['IMDB_Rating'].max()
    normalizedGroup = (group['IMDB_Rating'] - min) / (max - min)
    group['normalize_imdb_rating'] = normalizedGroup
    return group 

movies.groupby('Genre').apply(normalize_imdb_rating)
````
