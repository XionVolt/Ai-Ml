```python
# plotting a 2d plot
# x = y (means slope = 1 , and on increasing x , y is also increasing with same value,  same with decreasing x, y is also decreasing with same value)

import numpy as np
from matplotlib import pyplot as plt # or import matplotlib.pyplot as plt
x = np.linspace(0, 10, 2)
y = x
plt.figure(figsize=(7, 5))
plt.grid()
plt.plot(x, y) # means it plot from 0,0 to 10,10
```

```python
# y = x^2 (parabola)
x = np.linspace(-10, 10, 100)
y = x ** 2
plt.figure(figsize=(7, 5))
plt.grid()
plt.plot(x, y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# More precise x values
x = np.linspace(-10, 10, 100)  # Increased from 100 to 1000
y = 1 / (1 + np.exp(-x))

plt.figure(figsize=(7, 5))
plt.grid(True)
plt.plot(x, y, label="Sigmoid Curve", color="blue")

plt.title("Sigmoid Function")
plt.xlabel("x")
plt.ylabel("$\sigma$(x)")
plt.xticks(np.arange(-10, 11, 1))
plt.yticks(np.arange(0, 1.1, 0.1))
plt.legend()
plt.show()
```
