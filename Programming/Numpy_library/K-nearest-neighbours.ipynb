```python
import numpy as np 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# First we will make eucledian distance function
def eucledian_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))
```

```python
# Make 5 random points(basically you can say these are 5 random vectors in 3d space)
rng = np.random.default_rng(77)
vectors = rng.uniform(0, 1, (5, 3)) # use uniform distribution to generate random points
vectors
```

```python
# now we create 2 random classes for these vectors(let's say of setosa(red) and versicolor(blue))

rng = np.random.default_rng(7)
classes = rng.uniform(0, 2, 5).astype(int) # truncating the float values to int

classes = np.where(classes == 0, 'setosa', 'versicolor') # if class is 0 then setosa else versicolor
classes
```

```python
# plot the points
fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
for i, vector in enumerate(vectors):
    ax.quiver(0, 0, 0, vector[0], vector[1], vector[2], color='r' if classes[i] == 'setosa' else 'b')
```

```python
# Input vector
input_vector = np.array(list(map(float, input("Enter the input vector: ").split())))
print(input_vector) # color : green

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# -------------------------------------
# Previous cell code (because we updating graph)
for i, vector in enumerate(vectors):
    ax.quiver(0,0,0,vector[0], vector[1], vector[2], color='r' if classes[i] == 'setosa' else 'b')
# -------------------------------------

ax.quiver(0, 0, 0, input_vector[0], input_vector[1], input_vector[2], color='g')

plt.draw()
plt.show()
```

```python
distances = np.array([eucledian_distance(input_vector, vector) for vector in vectors])
min_distance_vector = np.argmin(distances)
print('minimum distanced vector from input vector is:', vectors[min_distance_vector] , 'that means our input vector belongs to class:', classes[min_distance_vector])
```

So we implemented a simplified K-Nearest Neighbors (KNN) classifier with $k = 1 (1-NN)$ (means we are using 1 nearest neighbour)

##  will 
- Pick k neighbors (instead of just 1).
- Use majority voting to classify the input vector.

- Apply regression:
   -  For that -  take the average of k nearest neighbors instead of classification
