```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from moviepy.editor import ImageSequenceClip
import os
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds 
```

```python
# scatter plot on animation -> https://youtu.be/MqppsRq3EJQ?si=7lnZ4-OciPB7Qd-Q&t=14857

# so basically it give us data points of specific category on specific frame
fig1 = px.scatter_3d(
    diamonds,
    x='carat',
    y='price',
    z='depth',
    animation_frame='color',
    color='color'
) 

fig1.update_layout(
    height=700,
    width=800,
    title=('carat, price and depth comparison by color')
)
fig1.show()
```
