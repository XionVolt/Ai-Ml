```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds
```

```python
# scatter plot with marginal histograms -> https://youtu.be/MqppsRq3EJQ?si=MdMB7ja9jxC184qg&t=14437

# so its like a jointgrid of seaborn, where marginal 
fig1 = px.scatter(diamonds, x="carat", y="price", marginal_x="box", marginal_y="histogram")
fig1.show()
```

```python
# we can also add facet_col 

# we not get marginal plot of y-axis because we give facet_col, if we give it facet_row it will give us marginal plot of y-axis instead of x-axis
fig2 = px.scatter(diamonds, x="carat", y="price", marginal_x="box", marginal_y="histogram",facet_col='color')
fig2.show()
```
