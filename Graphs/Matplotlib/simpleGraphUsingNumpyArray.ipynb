```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


# Define restaurant names
restaurant_names = ["Paradise Biryani", "Beijing Bites", "Pizza Hub", "Burger Point", "Chai Point"]

# Sales data (excluding restaurant_id column)
sales_data = np.array([
    [150000, 180000, 220000, 250000],  # Paradise Biryani
    [120000, 140000, 160000, 190000],  # Beijing Bites
    [200000, 230000, 260000, 300000],  # Pizza Hub
    [180000, 210000, 240000, 270000],  # Burger Point
    [160000, 185000, 205000, 230000]   # Chai Point
])

# Compute mean sales across years
mean_sales = np.mean(sales_data, axis=1)

# Create bar chart
plt.figure(figsize=(8, 6))
plt.bar(restaurant_names, mean_sales, color='skyblue', alpha=0.8)

# Plot a line connecting the top of the bars
plt.plot(restaurant_names, mean_sales, marker='o', color='red', linestyle='-', linewidth=2, markersize=6)

# Labels and title
plt.xlabel("Restaurant Names")
plt.ylabel("Mean Sales (in dollars)")
plt.title("Mean Sales of Restaurants Over the Years")
plt.xticks(rotation=20)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show plot
plt.show()
```
